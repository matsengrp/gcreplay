```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import pickle
```

### Most recent results directory

```python
results = glob.glob("../nextflow/results/*")[-1]
results
```

### Load GC 38 and GC 57

```python
tree_38 = pickle.load(open(f"{results}/gctrees/PR1.2-2-LA-38-GC/gctree.p", "rb"))
tree_57 = pickle.load(open(f"{results}/gctrees/PR1.8-3-LB-57-GC//gctree.p", "rb"))
```

### Compute LB stats on each tree

```python
tree_38.local_branching(tau=1, tau0=0.1)
tree_57.local_branching(tau=1, tau0=0.1)
```

### Plot with LBR colormap

```python
vmin=0
vmax=10
cmap="plasma"
```

```python
colormap = tree_38.feature_colormap("LBR", vmin=vmin, vmax=vmax, cmap=cmap)
tree_38.render("%%inline", colormap=colormap, idlabel=True)
```

```python
colormap = tree_57.feature_colormap("LBR", vmin=vmin, vmax=vmax, cmap=cmap)
tree_57.render("%%inline", colormap=colormap, idlabel=True)
```

### Loop over nodes and print LB info

```python
for node in tree_38.tree.traverse():
    print(f"{node.name}\t{node.LB_down:.2f}\t{node.LB_up:.2f}\t{node.LBR:.2f}")
```

```python
for node in tree_57.tree.traverse():
    print(f"{node.name}\t{node.LB_down:.2f}\t{node.LB_up:.2f}\t{node.LBR:.2f}")
```
